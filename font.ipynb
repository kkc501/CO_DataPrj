```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np

def example_f(x):
    return x**2

x = np.arange(0,100,1)
plt.xlabel('이것은 엑스축')
plt.ylabel('이것은 와이축')
plt.title('이것은 제목')
plt.plot(x, example_f(x))
```
